```python
import pandas 
import pandas as pd
```

```python
returns = pd.read_csv("amd_stock_data.csv")
```

```python
#Display summary statistics of the input data to get a feel for your data.
returns.describe(include="all")
```

```python
# Drop unneccessary columns (volume, open, high, and low)
returns.drop(columns=["open", "high", "volume", "low"], inplace =True)
returns.head()
```

```python
# Set the date column as the index
returns.set_index(pd.to_datetime(returns["date"], infer_datetime_format=True), inplace=True)
returns.head()
```

```python
# Drop the extra date column. (The index can now replace it.)
returns.drop(columns=["date"], inplace=True)
returns.head(10)
```

```python
# Calculate daily returns
daily_returns = returns.pct_change()
daily_returns.head()
```

```python
# Use loc() to select subsets from the DatetimeIndex to create date ranges of 1, 3, 5, and 10 years. loc[start_date:end_date].
subset = daily_returns.loc(['2018-04-30':'2019-04-29'])
```
